```python
import requests
from IPython.display import Image
from IPython.display import display

url = 'http://0.0.0.0:8000/semantic_search/'
```

```python
def view_images(text):
    myobj = {'input_text': text}
    response = requests.post(url, json = myobj)
    images_info = response.json()
    for image_info in images_info:
        image = Image(filename=f'Documents/work/{image_info["image"]}')
        display(image)
```

```python
view_images("cat")
```

```python
view_images("man")
```

```python
view_images("tree")
```
